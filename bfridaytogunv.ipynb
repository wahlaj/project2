```python
import pandas as pd
import numpy as np
import datetime as dt
from sqlalchemy import create_engine
```

```python
gunfile = "gun-violence-data_01-2013_03-2018.csv"
gun_df = pd.read_csv(gunfile)
gun_df.head()
```

```python
gun_df.count()
```

```python
clean_gun_df = gun_df.dropna(how='any')
clean_gun_df.head()
```

```python
clean_gun = clean_gun_df[['state', 'date', 'n_killed', 'n_injured']]
```

```python
clean_gun.head()
```

```python
new_gun_df = clean_gun.reset_index()
new_gun_df[0:].replace('', np.nan, inplace=True)
new_gun_df.head()
```

```python
indexed_gun_df = new_gun_df.drop(['index'], axis=1)
indexed_gun_df.head()
```

```python
new_gun_columns = indexed_gun_df.rename(columns={"state": "State", "date": "Black Friday", "n_killed": "Killed", "n_injured": "Injured"})
new_gun_columns.head()
```

```python
new_gun_columns.to_csv(r'C:\Users\wahla\project2\gun_stats_small.csv')
```
